```python
class Empleado:
    def __init__(self, nombre, sueldo):
        self.nombre = nombre
        self.sueldo = sueldo

    def mostrar_datos(self):
        print("Nombre:", self.nombre)
        print("Sueldo:", self.sueldo)

    def pagar_impuestos(self):
        if self.sueldo > 3000:
            print("Debe pagar impuestos")
        else:
            print("No debe pagar impuestos")

empleado1 = Empleado("Carlos", 2001)

empleado1.mostrar_datos()

empleado1.pagar_impuestos()
```

```python
class Punto:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def cuadrante(self):

        if self.x > 0 and self.y > 0:
            print("El punto está en el Primer Cuadrante")

        elif self.x < 0 and self.y > 0:
            print("El punto está en el Segundo Cuadrante")

        elif self.x < 0 and self.y < 0:
            print("El punto está en el Tercer Cuadrante")

        elif self.x > 0 and self.y < 0:
            print("El punto está en el Cuarto Cuadrante")

        else:
            print("El punto está sobre un eje")

punto1 = Punto(4, -3)

punto1.cuadrante()
```
