```python
#ejercicio 1
class CuentaBancaria:
    def __init__(self, titular: str, saldo_inicial: float = 0):

        self.__titular = titular
        self.__saldo   = 0.0
        if saldo_inicial > 0:
            self.depositar(saldo_inicial)

    @property
    def titular(self) -> str:
        return self.__titular

    @property
    def saldo(self) -> float:

        return self.__saldo


    def depositar(self, monto: float) -> str:

        if monto <= 0:
            raise ValueError(f"El monto debe ser positivo. Recibido: {monto}")
        self.__saldo += monto
        return f"Depósito de ${monto:,.2f} exitoso. Saldo: ${self.__saldo:,.2f}"

    def retirar(self, monto: float) -> str:

        if monto <= 0:
            raise ValueError(f"El monto debe ser positivo. Recibido: {monto}")
        if monto > self.__saldo:
            raise ValueError(
                f"Fondos insuficientes. Disponible: ${self.__saldo:,.2f}, "
                f"Solicitado: ${monto:,.2f}"
            )
        self.__saldo -= monto
        return f"Retiro de ${monto:,.2f} exitoso. Saldo: ${self.__saldo:,.2f}"

    def mostrar_saldo(self) -> str:

        return f"Titular: {self.__titular} | Saldo: ${self.__saldo:,.2f}"

    def __repr__(self) -> str:
        return f"CuentaBancaria(titular={self.__titular!r}, saldo={self.__saldo})"




if __name__ == "__main__":
    cuenta = CuentaBancaria("Ana Torres", saldo_inicial=500)

    print(cuenta.mostrar_saldo())


    print(cuenta.depositar(200))


    print(cuenta.retirar(150))

    try:
        cuenta.retirar(9999)
    except ValueError as e:
        print(f"Error: {e}")
```

```python
class Estudiante:
    def __init__(self, nombre, codigo):

        self.__nombre = nombre
        self.__codigo = codigo
        self.__notas = []


    def agregar_nota(self, nota):
        if 0 <= nota <= 5:
            self.__notas.append(nota)
            print("Nota agregada correctamente.")
        else:
            print("Error: la nota debe estar entre 0 y 5.")


    def calcular_promedio(self):
        if len(self.__notas) == 0:
            return 0
        return sum(self.__notas) / len(self.__notas)

    def mostrar_estado(self):
        promedio = self.calcular_promedio()

        print("\n--- DATOS DEL ESTUDIANTE ---")
        print(f"Nombre: {self.__nombre}")
        print(f"Código: {self.__codigo}")
        print(f"Notas: {self.__notas}")
        print(f"Promedio: {promedio:.2f}")

        if promedio >= 4:
            print("Estado: APROBADO")
        else:
            print("Estado: REPROBADO")


est1 = Estudiante("Andres", 101)

est1.agregar_nota(4.5)
est1.agregar_nota(3.8)
est1.agregar_nota(5)
est1.agregar_nota(6)

est1.mostrar_estado()
```

```python
# ejercicio 3

class Producto:
    def __init__(self, nombre, precio, stock):

        self.__nombre = nombre


        if precio > 0:
            self.__precio = precio
        else:
            self.__precio = 1

        if stock >= 0:
            self.__stock = stock
        else:
            self.__stock = 0

    def vender_producto(self, cantidad):
        if cantidad <= self.__stock:
            self.__stock -= cantidad
            print(f"Se vendieron {cantidad} unidades.")
        else:
            print("Error: stock insuficiente.")

    def agregar_stock(self, cantidad):
        if cantidad > 0:
            self.__stock += cantidad
            print(f"Se agregaron {cantidad} unidades al stock.")
        else:
            print("Error: la cantidad debe ser mayor a 0.")

    def cambiar_precio(self, nuevo_precio):
        if nuevo_precio > 0:
            self.__precio = nuevo_precio
            print("Precio actualizado correctamente.")
        else:
            print("Error: el precio debe ser mayor a 0.")

    def mostrar_info(self):
        print("\n--- INFORMACIÓN DEL PRODUCTO ---")
        print(f"Nombre: {self.__nombre}")
        print(f"Precio: ${self.__precio}")
        print(f"Stock disponible: {self.__stock}")



p1 = Producto("Laptop", 2500, 10)

p1.mostrar_info()

p1.vender_producto(3)
p1.agregar_stock(5)
p1.cambiar_precio(3000)

p1.mostrar_info()

p1.vender_producto(20)
p1.cambiar_precio(-50)
```

```python
#ejercicio 4
class Empleado:
    def __init__(self, nombre, salario, horas_trabajadas):

        self.__nombre = nombre


        if salario > 0:
            self.__salario = salario
        else:
            self.__salario = 1


        if horas_trabajadas >= 0:
            self.__horas_trabajadas = horas_trabajadas
        else:
            self.__horas_trabajadas = 0


    def calcular_pago(self):
        pago = self.__salario * self.__horas_trabajadas
        return pago


    def aumentar_salario(self, aumento):
        if aumento > 0:
            self.__salario += aumento
            print("Salario aumentado correctamente.")
        else:
            print("Error: el aumento debe ser mayor a 0.")


    def mostrar_informacion(self):
        print("\n--- INFORMACIÓN DEL EMPLEADO ---")
        print(f"Nombre: {self.__nombre}")
        print(f"Salario por hora: ${self.__salario}")
        print(f"Horas trabajadas: {self.__horas_trabajadas}")
        print(f"Pago total: ${self.calcular_pago()}")



emp1 = Empleado("Andres", 20000, 40)

emp1.mostrar_informacion()

emp1.aumentar_salario(5000)

emp1.mostrar_informacion()

```

```python
#Ejercicio 6

class Libro:
    def __init__(self, titulo, autor):

        self.__titulo = titulo
        self.__autor = autor
        self.__disponible = True


    def prestar(self):
        if self.__disponible:
            self.__disponible = False
            print(f'El libro "{self.__titulo}" ha sido prestado.')
        else:
            print(f'El libro "{self.__titulo}" no está disponible.')


    def devolver(self):
        if not self.__disponible:
            self.__disponible = True
            print(f'El libro "{self.__titulo}" ha sido devuelto.')
        else:
            print(f'El libro "{self.__titulo}" ya estaba disponible.')


    def mostrar_estado(self):
        print("\n--- INFORMACIÓN DEL LIBRO ---")
        print(f"Título: {self.__titulo}")
        print(f"Autor: {self.__autor}")

        if self.__disponible:
            print("Estado: Disponible")
        else:
            print("Estado: Prestado")



libro1 = Libro("Cien Años de Soledad", "Gabriel García Márquez")

libro1.mostrar_estado()

libro1.prestar()

libro1.mostrar_estado()

libro1.prestar()

libro1.devolver()

libro1.mostrar_estado()

libro1.devolver()
```

```python
#Ejercicio 8

class HabitacionHotel:
    def __init__(self, numero, precio_noche):

        self.__numero = numero


        if precio_noche > 0:
            self.__precio_noche = precio_noche
        else:
            self.__precio_noche = 1

        self.__ocupada = False


    def reservar(self):
        if not self.__ocupada:
            self.__ocupada = True
            print(f"Habitación {self.__numero} reservada correctamente.")
        else:
            print(f"La habitación {self.__numero} ya está ocupada.")


    def liberar(self):
        if self.__ocupada:
            self.__ocupada = False
            print(f"Habitación {self.__numero} liberada.")
        else:
            print(f"La habitación {self.__numero} ya está disponible.")


    def calcular_total(self, noches):
        if noches > 0:
            total = self.__precio_noche * noches
            return total
        else:
            print("Error: el número de noches debe ser mayor a 0.")
            return 0


    def mostrar_estado(self):
        print("\n--- INFORMACIÓN DE LA HABITACIÓN ---")
        print(f"Número: {self.__numero}")
        print(f"Precio por noche: ${self.__precio_noche}")

        if self.__ocupada:
            print("Estado: Ocupada")
        else:
            print("Estado: Disponible")



h1 = HabitacionHotel(101, 150000)

h1.mostrar_estado()

h1.reservar()

h1.mostrar_estado()


noches = 3
print(f"\nCosto total por {noches} noches: ${h1.calcular_total(noches)}")

h1.liberar()

h1.mostrar_estado()

h1.reservar()
h1.reservar()
```

```python
#Ejercicio 9

class CajeroAutomatico:
    def __init__(self, saldo, pin):

        self.__saldo = saldo
        self.__pin = pin
        self.__intentos = 0
        self.__bloqueado = False


    @property
    def saldo(self):
        return self.__saldo


    @saldo.setter
    def saldo(self, nuevo_saldo):
        if nuevo_saldo >= 0:
            self.__saldo = nuevo_saldo
        else:
            raise ValueError("El saldo no puede ser negativo.")


    def autenticar(self, pin):
        if self.__bloqueado:
            print("Cuenta bloqueada por demasiados intentos fallidos.")
            return False

        if pin == self.__pin:
            print("Autenticación exitosa.")
            self.__intentos = 0
            return True
        else:
            self.__intentos += 1
            print(f"PIN incorrecto. Intento {self.__intentos}/3")

            if self.__intentos >= 3:
                self.__bloqueado = True
                print("Cuenta bloqueada.")

            return False


    def retirar(self, monto):
        try:
            if monto <= 0:
                raise ValueError("El monto debe ser mayor a 0.")

            if monto > self.__saldo:
                raise ValueError("Saldo insuficiente.")

            self.__saldo -= monto
            print(f"Retiro exitoso. Nuevo saldo: ${self.__saldo}")

        except ValueError as e:
            print("Error:", e)


    def depositar(self, monto):
        try:
            if monto <= 0:
                raise ValueError("El depósito debe ser mayor a 0.")

            self.__saldo += monto
            print(f"Depósito exitoso. Nuevo saldo: ${self.__saldo}")

        except ValueError as e:
            print("Error:", e)


    def consultar_saldo(self):
        print(f"Saldo actual: ${self.__saldo}")



cajero = CajeroAutomatico(500000, "1234")


cajero.autenticar("1111")
cajero.autenticar("2222")
cajero.autenticar("3333")


cajero.autenticar("1234")

print("\n--- Nuevo Cajero ---")

cajero2 = CajeroAutomatico(300000, "4321")

if cajero2.autenticar("4321"):
    cajero2.consultar_saldo()
    cajero2.depositar(50000)
    cajero2.retirar(100000)
    cajero2.consultar_saldo()
```

```python
#Ejercicio 10

class SistemaAcademico:
    def __init__(self):

        self.__estudiantes = {}
        self.__cursos = []
        self.__calificaciones = {}


    @property
    def estudiantes(self):
        return self.__estudiantes


    @property
    def cursos(self):
        return self.__cursos


    def registrar_estudiante(self, codigo, nombre):
        try:
            if codigo in self.__estudiantes:
                raise ValueError("El estudiante ya está registrado.")

            self.__estudiantes[codigo] = nombre
            self.__calificaciones[codigo] = {}

            print(f"Estudiante {nombre} registrado correctamente.")

        except ValueError as e:
            print("Error:", e)


    def registrar_curso(self, curso):
        try:
            if curso in self.__cursos:
                raise ValueError("El curso ya existe.")

            self.__cursos.append(curso)
            print(f"Curso '{curso}' registrado correctamente.")

        except ValueError as e:
            print("Error:", e)


    def asignar_calificacion(self, codigo, curso, nota):
        try:
            if codigo not in self.__estudiantes:
                raise ValueError("El estudiante no existe.")

            if curso not in self.__cursos:
                raise ValueError("El curso no existe.")

            if nota < 0 or nota > 5:
                raise ValueError("La nota debe estar entre 0 y 5.")

            self.__calificaciones[codigo][curso] = nota

            print("Calificación asignada correctamente.")

        except ValueError as e:
            print("Error:", e)


    def generar_reporte(self):
        print("\n====== REPORTE ACADÉMICO ======")

        for codigo, nombre in self.__estudiantes.items():
            print(f"\nEstudiante: {nombre}")
            print(f"Código: {codigo}")

            notas = self.__calificaciones[codigo]

            if len(notas) == 0:
                print("No tiene calificaciones registradas.")
            else:
                suma = 0

                for curso, nota in notas.items():
                    print(f"{curso}: {nota}")
                    suma += nota

                promedio = suma / len(notas)

                print(f"Promedio: {promedio:.2f}")

                if promedio >= 3:
                    print("Estado: APROBADO")
                else:
                    print("Estado: REPROBADO")



sistema = SistemaAcademico()


sistema.registrar_estudiante(101, "Andres")
sistema.registrar_estudiante(102, "Maria")


sistema.registrar_curso("Matemáticas")
sistema.registrar_curso("Programación")


sistema.asignar_calificacion(101, "Matemáticas", 4.5)
sistema.asignar_calificacion(101, "Programación", 3.8)

sistema.asignar_calificacion(102, "Matemáticas", 2.5)
sistema.asignar_calificacion(102, "Programación", 2.9)


sistema.asignar_calificacion(101, "Programación", 7)


sistema.generar_reporte()
```
