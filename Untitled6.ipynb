```python
"""Ejercicio 1 — Herencia Básica de Vehículos
Clase padre: Vehiculo.
Atributos: marca y color.
Métodos: acelerar() y frenar().
Clase hija: Carro con atributo puertas y método tocar_bocina(). """


class vehiculos:
  def __init__(self,marca,color):
    self.marca = marca
    self.color = color

  def acelerar(self):
    print("el vehiculo esta accelerando...")

  def frenar(self):
    print("el vehiculo esta frenando...")

class carro(vehiculos):
  def __init__(self, marca, color,puerta = int):
    super().__init__(marca, color)
    self.puerta = puerta

  def tocar_bocina(self):
    print("sonido nde bocina...")

car = carro("ferrary","negro", 4)
print(car.marca)
car.tocar_bocina()
car.acelerar()
```

```python
"""Ejercicio 2 — Sistema de Animales
Clase padre: Animal.
Atributos: nombre y edad.
Método: hacer_sonido().
Clases hijas: Perro y Gato con métodos propios. """

class Animal:
  def __init__(self,nombre, edad):
    self.nombre = nombre
    self.edad = edad

  def hacer_sonido(self):
    print("sonido de animal...")

class perro(Animal):
  def __init__(self, nombre, edad, color):
    super().__init__(nombre, edad)
    self.color = color

  def correr(self,nombre):
    print(f"el perro: {nombre}, esta corriendo muy rapido...")

class gato(Animal):
  def __init__(self, nombre, edad, color):
    super().__init__(nombre, edad)
    self.color = color

  def trepar(self,nombre):
    print(f"el gato: {nombre}, trepa muy alto...")

dog = perro("firulais","12","negro")
cat = gato("michi","3","blanco con negro")
print(f"nombre: {dog.nombre}")
dog.correr(dog.nombre)
dog.hacer_sonido()
print(f"nombre: {cat.nombre}")
cat.trepar(cat.nombre)
cat.hacer_sonido()
```

```python
""" Ejercicio 3 — Sistema Académico
Clase padre: Persona.
Atributos: nombre y documento.
Método: mostrar_datos().
Clases hijas: Estudiante y Profesor."""

class persona:
  def __init__(self,nombre, documento = int):
    self.nombre = nombre
    self.documento = documento

  def mostrar_datos(self,nombre,documento):
    print(f"""
    ___ Datos____
    Nombre: {nombre}
    Documento: {documento}""")

class estudiante(persona):
  pass

class Profesor(persona):
  pass

datos1 = estudiante("juan",23232434)
datos2 = Profesor("sancho",121218328)

datos1.mostrar_datos(datos1.nombre ,  datos1.documento)
datos2.mostrar_datos(datos2.nombre ,  datos2.documento)
```

```python
"""Ejercicio 4 — Empleados de una Empresa
Clase padre: Empleado.
Atributos: nombre y salario.
Método: calcular_pago().
Clases hijas: EmpleadoTiempoCompleto y EmpleadoPorHoras. """

class Empleado:
    def __init__(self, nombre, salario):
        self.__nombre = nombre
        self.__salario = salario

    @property
    def salario(self):
        return self.__salario

    @property
    def nombre(self):
        return self.__nombre

    def calcular_pago(self):
        return self.__salario


class EmpleadoTiempoCompleto(Empleado):
    def calcular_pago(self):
        return self.salario + 500


class EmpleadoPorHoras(Empleado):
    def __init__(self, nombre, salario, horas):
        super().__init__(nombre, salario)
        self.__horas = horas

    def calcular_pago(self):
        return self.salario * self.__horas


emp1 = EmpleadoTiempoCompleto("Luis", 2000)
emp2 = EmpleadoPorHoras("Ana", 20, 40)

print(emp1.calcular_pago())
print(emp2.calcular_pago())
```

```python
""" Ejercicio 5 — Sistema Bancario
Clase padre: CuentaBancaria.
Atributos: titular y saldo.
Métodos: depositar() y retirar().
Clases hijas: CuentaAhorros y CuentaCorriente."""


class CuentaBancaria:
    def __init__(self, titular, saldo):
        self.__titular = titular
        self.__saldo = saldo

    @property
    def saldo(self):
        return self.__saldo

    def depositar(self, monto):
        if monto > 0:
            self.__saldo += monto
            print("Depósito realizado.")

    def retirar(self, monto):
        if monto <= self.__saldo:
            self.__saldo -= monto
            print("Retiro exitoso.")
        else:
            print("Saldo insuficiente.")


class CuentaAhorros(CuentaBancaria):
    def retirar(self, monto):
        print("Retiro desde cuenta de ahorros:")
        super().retirar(monto)


class CuentaCorriente(CuentaBancaria):
    def retirar(self, monto):
        print("Retiro desde cuenta corriente:")
        super().retirar(monto)


cuenta = CuentaAhorros("Andres", 5000)
cuenta.depositar(1000)
cuenta.retirar(2000)
print(cuenta.saldo)
```

```python
""" Ejercicio 6 — Videojuego de Personajes
Clase padre: Personaje.
Atributos: nombre, vida y ataque.
Métodos: atacar() y defender().
Clases hijas: Guerrero, Mago y Arquero."""

class Personaje:
    def __init__(self, nombre, vida, ataque):
        self.__nombre = nombre
        self.__vida = vida
        self.__ataque = ataque

    @property
    def nombre(self):
        return self.__nombre

    def atacar(self):
        print(f"{self.__nombre} ataca con {self.__ataque} puntos.")

    def defender(self):
        print(f"{self.__nombre} se está defendiendo.")


class Guerrero(Personaje):
    def atacar(self):
        print(f"{self.nombre} usa espada poderosa.")


class Mago(Personaje):
    def atacar(self):
        print(f"{self.nombre} lanza un hechizo mágico.")


class Arquero(Personaje):
    def atacar(self):
        print(f"{self.nombre} dispara una flecha.")


personajes = (Guerrero("Thor", 100, 30),Mago("Merlin", 80, 50), Arquero("Legolas", 90, 40))

for p in personajes:
    p.atacar()
```

```python
""" Clase padre: Transporte.
Atributos: capacidad y velocidad.
Método: mover().
Clases hijas: Avion, Barco y Tren."""


class Transporte:
  def __init__(self,capacidad, velocidad):
    self.capacidad = capacidad
    self.velocidad = velocidad

  def mover(self):
    print("Arrancando...")

class avion(Transporte):
  pass

class Barco(Transporte):
  pass

class Tren(Transporte):
  pass

tran1 = avion(30,533)
tran2 = Barco(10,542)
tran3 = Tren(65,5355)

print("Avion: ")
tran1.mover()

print("Barco: ")
tran2.mover()

print("Tren: ")
tran3.mover()
```

```python
"""Ejercicio 8 — Sistema de Tienda Online
Clase padre: Producto.
Atributos: nombre y precio.
Método: calcular_descuento().
Clases hijas: Electronico, Ropa y Alimento. """

class Producto:
    def __init__(self, nombre, precio):
        self.__nombre = nombre
        self.__precio = precio

    @property
    def precio(self):
        return self.__precio

    @property
    def nombre(self):
        return self.__nombre

    def calcular_descuento(self):
        return self.__precio

class Electronico(Producto):
  def calcular_descuento(self):
    return self.precio * 0.90


class Ropa(Producto):
  def calcular_descuento(self):
    return self.precio * 0.80


class Alimento(Producto):
  def calcular_descuento(self):
    return self.precio * 0.95


productos = (Electronico("Laptop", 3000),Ropa("Camisa", 100), Alimento("Pan", 20))

for p in productos:
    print(f"{p.nombre}: {p.calcular_descuento()}")
```

```python
"""Ejercicio 9 — Sistema Hospitalario
Clase padre: EmpleadoHospital.
Atributos: nombre y turno.
Método: trabajar().
Clases hijas: Doctor, Enfermero y Administrador."""

class EmpleadoHospital:
  def __init__(self, nombre, turno):
    self.__nombre = nombre
    self.__turno = turno

  @property
  def nombre(self):
    return self.__nombre

  def trabajar(self):
    print(f"{self.__nombre} está trabajando.")

class Doctor(EmpleadoHospital):
  def trabajar(self):
    print(f"El doctor {self.nombre} atiende pacientes.")

class Enfermero(EmpleadoHospital):
  def trabajar(self):
    print(f"El enfermero {self.nombre} ayuda en emergencias.")

class Administrador(EmpleadoHospital):
  def trabajar(self):
    print(f"El administrador {self.nombre} organiza documentos.")

empleados = (Doctor("Juan", "Mañana"), Enfermero("Pedro", "Tarde"), Administrador("Laura", "Noche"))

for e in empleados:
    e.trabajar()
```
